## What is survival analysis?

Statistical methods to analyze data with **time to event** outcomes

* **Survival**: Probability of survival
* **Hazard**: Rate of "death" (or event occurrence)
* **Regression**: Relationship of factors with event outcome

## Lifelines' features

```python
# Demonstrated on Major League Soccer draft data (1996-2015)
draft_data.sample(10)
```

### Survival curves

```python
import lifelines

top_qtr = draft_data[draft_data.draft_pick < 0.25]

kmf = lifelines.KaplanMeierFitter()
kmf.fit(top_qtr["matches"], event_observed=top_qtr["inactive"], label='Top 25% of draft')
kmf.plot()
plt.xlabel('MLS Match Appearances', fontweight='bold')
plt.ylabel('Proportion of Draftees Remaining', fontweight='bold')
```

### Hazard curves

```python
naf = lifelines.NelsonAalenFitter()
naf.fit(top_qtr["matches"], event_observed=top_qtr["inactive"], label='Top 25% of draft')
naf.plot()
plt.xlabel('MLS Match Appearances', fontweight='bold')
plt.ylabel('Cumulative Hazard', fontweight='bold')
```
